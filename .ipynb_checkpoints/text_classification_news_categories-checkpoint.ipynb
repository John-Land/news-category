# News Category Text Classfication

One popular text classification application is used in digital advertising for contextual targeting. <br>
For contextual targeting, ads are placed based on the context of the news article, assuming that ads placed in a related context have a higher probability of being clicked on (higher click throuh rate). <br>

In this project we train several deep learning models to predict the news topic of an article from the raw text of the article.

# Load and Pre-Process Data

This dataset used contains around 200k news headlines from the year 2012 to 2018 obtained from HuffPost. <br>
The raw dataset was downloaded from Kaggle. <br>

Refer to below link for more details:
<https://www.kaggle.com/rmisra/news-category-dataset> <br>

For model training, we will use the combined text from the headline + short description, to try to predice the news category out of 41 different categories.

```python
#dependancies

#data analysis and scientific computing
import numpy as np 
import pandas as pd

#data visualization
import matplotlib.pyplot as plt
import seaborn as sns

#machine learning and Deep Learning
import tensorflow as tf 
from tensorflow import keras 
```

```python
#load data from file
path = 'data/'
file = 'News_Category_Dataset_v2.json'

news_data_orig = pd.read_json(path+file, lines=True)
news_data = news_data_orig.copy()
news_data
```

```python
#check data for missing values
news_data.isna().sum()
```

```python
#split data into predictor (headline + short description) and label (news category)
X = news_data['headline'] + '. ' + news_data['short_description'].copy()
y = news_data['category'].copy()
```

```python
X.head()
```

```python
y.head()
```

```python
#encode categorical target valriable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder().fit(y)
y = pd.Series(le.transform(y))
y.head()
```

```python
#initialize label dictionary to convert label integer to news category
label_dict = dict(enumerate(le.classes_))
label_dict
```

```python
#split the dataset into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=0, shuffle=True, stratify=y)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)

#reset index
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

# Model Training and Evaluation

## Model 1: Pre-trained sentence embeddings (nnlm-en-dim50) + Fully Connected Layers

```python
#load layers and pre-trained model

#load keras model to initialize model
from keras import Model

#load layers required for model and regularization
from keras.layers import Input
from tensorflow_hub import KerasLayer
from keras.layers import Dropout
from keras.layers import Dense
from keras.regularizers import l1_l2

#from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import plot_model
```

```python
#define model

#input layer
input_text = Input(shape=(), dtype=tf.string, name='Input_Layer')

#get pre-trained word embeddings
hub_layer = KerasLayer("https://tfhub.dev/google/nnlm-en-dim50/2",input_shape=[], dtype=tf.string, trainable=False, name='Pre_Trained_Sentence_Embedding_Layer')(input_text)

#dense layer
regularizer = l1_l2(l1=0.001, l2=0.001)
dense_01 = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_01', kernel_regularizer=regularizer)(hub_layer)

#dropout layer
dropout_01 = Dropout(0.5, name='Dropout_Layer_01')(dense_01)

#dense layer
regularizer = l1_l2(l1=0.001, l2=0.001)
dense_02 = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_02', kernel_regularizer=regularizer)(dropout_01)

#dropout layer
dropout_02 = Dropout(0.5, name='Dropout_Layer_02')(dense_02)

#dense layer
regularizer = l1_l2(l1=0.001, l2=0.001)
dense_03 = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_03', kernel_regularizer=regularizer)(dropout_02)

#dropout layer
dropout_03 = Dropout(0.5, name='Dropout_Layer_03')(dense_03)

#dense output layer
regularizer = l1_l2(l1=0.001, l2=0.001)
output = Dense(units=len(np.unique(y_train)), activation='sigmoid', name='Dense_Output_Layer', kernel_regularizer=regularizer)(dropout_03)

#initialize model
model_01 = Model(inputs=input_text, outputs=output)
```

```python
model_01.summary()
```

```python
#plot model
plot_model(model = model_01, to_file='model_01.png')
```

```python
#show word embedding from tensorflow hub model
layer_name = model_01.layers[1].name
intermediate_layer_model = keras.Model(inputs=model_01.input,
                                       outputs=model_01.get_layer(layer_name).output)
intermediate_output = intermediate_layer_model(X_train)
print('Pre-trained word embedding from tensorflow hub model')
intermediate_output[0]
```

```python
# compile the model (should be done *after* setting layers to non-trainable)

#load optimization function
from tensorflow.keras.optimizers import Adam
optimization_function = Adam(learning_rate=0.001,
                             beta_1=0.9,
                             beta_2=0.999,
                             epsilon=1e-07,
                             amsgrad=False,
                             name="Adam")

#load loss function
from keras.losses import SparseCategoricalCrossentropy #as labels not one-hot encoded
loss_function = SparseCategoricalCrossentropy(from_logits=False)

#compile model with definined optimization and loss function
model_01.compile(optimizer=optimization_function, loss=loss_function, metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
callback_early_stopping = EarlyStopping(patience=10)

history_01 = model_01.fit(
                    x=X_train,
                    y=y_train,
                    batch_size=128,
                    epochs=500,
                    validation_split=0.1,
                    shuffle=True,
                    workers=-1,
                    use_multiprocessing=True,
                    callbacks = callback_early_stopping
                    )
```

```python
def plot_model_training(training_history, metric):
    '''
    Function plots the desiered metric over training epocs on both the training and validation set
    
    Parameters
    training_history = history of training data. output from tensorflow from model fitting
    metic = desired metric to plot over epocs. 'loss' for training and validation loss. 'accuracy' for training and validation accuracy
    
    '''
  
    fig, ax = plt.subplots(figsize=(10,6))
    ax.plot(training_history.history[metric], label='training_'+metric)
    ax.plot(training_history.history['val_' + metric], ls ='dashed', label='validation_'+metric)
    ax.set_title('Train and validation ' + metric + ' per training epoch')
    ax.set_xlabel('epoch')
    ax.set_ylabel(metric)
    ax.legend()
```

```python
plot_model_training(training_history=history_01, metric='loss')
```

```python
plot_model_training(training_history=history_01, metric='accuracy')
```

## Model 3: Pre-trained sentence embeddings (BERT) + Fully Connected Layers

```python
#load layers and pre-trained model

#load keras model to initialize model
from keras import Model

#load layers required for model and regularization
from keras.layers import Input
from tensorflow_hub import KerasLayer
from keras.layers import Dropout
from keras.layers import Dense
from keras.regularizers import l1_l2

#from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import plot_model

#pre-requisite for bert tensorflow hub model
import tensorflow_text as text
```

```python
#define model

#input layer
input_text = Input(shape=(), dtype=tf.string, name='Input_Layer')

#get pre-trained word embeddings
preprocess_layer = KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3", name='Pre_Trained_Embedding_Preprocess_Layer')(input_text)
embedding_layer = KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4", name='Pre_Trained_Sentence_Embedding_Layer', trainable=False)(preprocess_layer)['pooled_output']

#dense layer
regularizer = l1_l2(l1=0.001, l2=0.001)
dense_01 = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_01', kernel_regularizer=regularizer)(embedding_layer)

#dropout layer
dropout_01 = Dropout(0.5, name='Dropout_Layer_01')(dense_01)

#dense layer
regularizer = l1_l2(l1=0.001, l2=0.001)
dense_02 = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_02', kernel_regularizer=regularizer)(dropout_01)

#dropout layer
dropout_02 = Dropout(0.5, name='Dropout_Layer_02')(dense_02)

#dense layer
regularizer = l1_l2(l1=0.001, l2=0.001)
dense_03 = Dense(units=100, activation='relu', name='Dense_Hidden_Layer_03', kernel_regularizer=regularizer)(dropout_02)

#dropout layer
dropout_03 = Dropout(0.5, name='Dropout_Layer_03')(dense_03)

#dense output layer
regularizer = l1_l2(l1=0.001, l2=0.001)
output = Dense(units=len(np.unique(y_train)), activation='sigmoid', name='Dense_Output_Layer', kernel_regularizer=regularizer)(dropout_03)

#initialize model
model_03 = Model(inputs=input_text, outputs=output)
```

## Model 4: Pre-trained word embeddings (BERT) + RNN (GRU) layers

```python
#load layers and pre-trained model

#load keras model to initialize model
from keras import Model

#load layers required for model and regularization
from keras.layers import Input
from tensorflow_hub import KerasLayer
from keras.layers import GRU
from keras.layers import Dropout
from keras.layers import Dense
from keras.regularizers import l1_l2

#from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import plot_model

#pre-requisite for bert tensorflow hub model
import tensorflow_text as text
```

```python
#define model

#input layer
input_text = Input(shape=(), dtype=tf.string, name='Input_Layer')

#get pre-trained word embeddings
preprocess_layer = KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3", name='Pre_Trained_Embedding_Preprocess_Layer')(input_text)
embedding_layer = KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4", name='Pre_Trained_Word_Embedding_Layer', trainable=False)(preprocess_layer)['sequence_output']

#GRU Layer
GRU_01 = GRU(units=128, return_sequences=True, name='GRU_Layer_01')(embedding_layer)

#GRU Layer
GRU_02 = GRU(units=128, return_sequences=True, name='GRU_Layer_02')(GRU_01)

#GRU Layer
GRU_03 = GRU(units=128, return_sequences=False, name='GRU_Layer_03')(GRU_02)

#dense output layer
regularizer = l1_l2(l1=0.001, l2=0.001)
output = Dense(units=len(np.unique(y_train)), activation='sigmoid', name='Dense_Output_Layer', kernel_regularizer=regularizer)(GRU_03)

#initialize model
model_04 = Model(inputs=input_text, outputs=output)
```

# Best Model Test Set Performance

```python
#model evaluation metrics
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

#data visualization
import seaborn as sns

#define model performance function
def plot_model_performance(y_test, y_hat_test):
    '''
    plot model performance metrics and confusion matrix
    '''
    #print performance metrics on test set
    print('Test set Accuracy:', accuracy_score(y_true=y_test, y_pred=y_hat_test).round(3))
    print('Test set Precision:', precision_score(y_true=y_test, y_pred=y_hat_test, average='macro').round(3))
    print('Test set Recall:', recall_score(y_true=y_test, y_pred=y_hat_test, average='macro').round(3))
    
    #plot confusion matrix
    conf_matrix = confusion_matrix(y_true=y_test, y_pred=y_hat_test)
    fig, ax = plt.subplots(figsize=(25,10))
    ax = sns.heatmap(conf_matrix, annot=True, fmt="d", cmap='Blues');
    ax.set_title('Test set Confusion Matrix', fontsize = 20, fontweight ='bold')
    ax.set_xlabel('Predicted Class', fontweight ='bold', fontsize=15) 
    ax.set_ylabel('True Class', fontweight ='bold', fontsize=15)
```

```python
#get model predictions
y_hat_test_prob = model_04.predict(X_test)
y_hat_test = np.argmax(y_hat_test_prob, axis=1)

#plot model performance
plot_model_performance(y_test, y_hat_test)
plt.savefig('confusion_matrix_test_set')
```
